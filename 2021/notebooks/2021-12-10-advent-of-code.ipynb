```python
aoc_day = '2021-12-10'
test_input_file = f"../test_inputs/{aoc_day}-input.txt"
input_file = f"../inputs/{aoc_day}-input.txt"
try_test_input = False


def preprocess_input(input:str) -> list:
    input = [x for x in input.split("\n") if x != '']
    return input 

with open(test_input_file, 'r') as f:
    test_input = preprocess_input(f.read())

with open(input_file, 'r') as f:
    input = preprocess_input(f.read())

if try_test_input:
    input = test_input.copy()

input
```

```python
opening_chars = ['(', '[', '{', '<']
closing_chars = [')', ']', '}', '>']
chars_pairs_start_to_end = {k:v for k,v in zip(opening_chars, closing_chars)}
p1_scores = [3,57,1197,25137]
p1_scores_dict = {k:v for k,v in zip(closing_chars, p1_scores)}
p2_scores = [1,2,3,4]
p2_scores_dict = {k:v for k,v in zip(closing_chars, p2_scores)}

def calc_score(x):
  if x in closing_chars:
    return p1_scores_dict[x]
  else:
    return 0

def calc_score_p2(chars):
  char_list = [p2_scores_dict[x] for x in chars if x in closing_chars]
  score = 0 
  for x in char_list:
    score = score * 5
    score = score + x
  return score


def is_matching(x,y):
  if chars_pairs_start_to_end[x] == y:
    return True
  else:
    return False

def find_chunk(chars, char_start=0):
  chars_list = [x for x in chars]
  open_char_count = 0
  close_char_count = 0
  char_start = 0
  char = char_start
  opening_chars_list = []
  for c in chars_list:
    if c in opening_chars: 
      open_char_count+= 1
      opening_chars_list += [c]
    elif c in closing_chars:
      close_char_count+= 1
      # Chunk is corrupted
      last_opening_char = opening_chars_list[-1]
      if (c == chars_pairs_start_to_end[last_opening_char]):
        # Remove last character in list since its now closed 
        opening_chars_list = opening_chars_list[:-1]
      else:
        # print(f'Expected {chars_pairs_start_to_end[last_opening_char]} but got {c} at index {char}')
        return c

    # Chunk Ends
    if open_char_count == close_char_count:
      break
    
    char+=1
  return [opening_chars_list,char]
```

```python
# Part One
score = 0
for line in input:
  score += calc_score(find_chunk(line))

print(score)
```

```python
# Part Two

score_list = [] 

for line in input:
  parsed_line = find_chunk(line) 

  #Corrupted
  if parsed_line in closing_chars:
    continue

  print("\n\n")
  print(line)
  print(parsed_line)

  if parsed_line[0] == []:
    parsed_line = find_chunk(line[parsed_line[1]+1:])
    print(parsed_line)

  autocomplete = ''.join([chars_pairs_start_to_end[x] for x in parsed_line[0]][::-1])
  score = calc_score_p2(autocomplete)
  print(score)
  score_list += [score]

score_list.sort()
middle_index = round(len(score_list) / 2)
middle_score = score_list[middle_index]
print(middle_score)
```
